```python
import sys
import glob
import h5py
import cv2
import numpy as np  
import matplotlib.pyplot as plt  
```

```python
nfiles = len(glob.glob('./*.jpg'))
print(f'Reading dir: nfiles={nfiles}')

# load each image into a separate dataset (image NOT resized)    
with h5py.File('mytestfile.hdf5','w') as  h5f:
    grp = h5f.create_group("bar")
    subgrp = grp.create_group("baz")
    for cnt, ifile in enumerate(glob.iglob('./*.jpg')) :
        img = cv2.imread(ifile, cv2.IMREAD_COLOR)
        # or use cv2.IMREAD_GRAYSCALE, cv2.IMREAD_UNCHANGED
        img_ds = subgrp.create_dataset('images_'+f'{cnt+1:03}', data=img)
```
